# The NFW profile

The Navarro-Frenk-White (NFW) profile the the dark matter halo density profile of choice in Tinker and Chen's paper. The problem is to get the concentration parameter right

```python
from halotools.empirical_models import NFWProfile
from astropy.units.astrophys import Msun, Mpc
import astropy.units as u
from astropy.cosmology import Planck13

import numpy as np
import matplotlib.pyplot as plt
```

## Concentration model

```python
masses = 10**np.linspace(10,14)/Planck13.h
for z in np.linspace(0,2.5,6):
    nfw = NFWProfile(cosmology=Planck13,redshift=z,conc_mass_model='dutton_maccio14',mdef='200m')
    conc = nfw.conc_NFWmodel(prim_haloprop=masses)
    plt.plot(np.log10(masses),np.log10(conc),label="z = {:1.1f}".format(z))
plt.legend()
plt.ylabel("$\log c_{200}$")
plt.xlabel("$\log M/h^{-1}M_\odot$")
plt.title("Concentration mass relation: Dutton-Maccio 2014")
plt.show()
```
